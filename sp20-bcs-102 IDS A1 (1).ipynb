```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
dataset = pd.read_csv('dataset.csv')
print(dataset)

```

```python
#Print the list of all students whose first name starts with letter the 'H'.
name = dataset[dataset.Name.str.contains('^H')]
print(name.Name)
```

```python
#Print the percentage of students who have a CGPA of 3.0 or above.
CGPA=dataset.CGPA
GPA = dataset[dataset.CGPA.str.contains('^3')]
gett=GPA.CGPA
count=0
count2=0
for above in gett:
    count+=1  
for total in CGPA:
     count2+=1
percentage=(count*100) / count2
print(percentage)
```

```python
#Plot a pie chart to show the ratio of male and female students
male = dataset[dataset.Gender.str.contains('^M')]
get_male=male.Gender

count_male=0
count_female=0

for x in get_male:
    count_male+=1  
   
    
female = dataset[dataset.Gender.str.contains('^F')]
get_female=female.Gender

for y in get_female:
    count_female+=1  

y=np.array([count_male,count_female])
 
label=['Male','Female']

plt.pie(y,labels=label, autopct='%1.2f%%')
plt.legend()
plt.show()

```
